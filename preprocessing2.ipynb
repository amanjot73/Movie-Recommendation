```python
import pandas as pd
# chunk_size = 1000
# chunks = pd.read_csv(
#     'title.principals.tsv',
#     sep='\t',
#     na_values='\\N',
#     chunksize=chunk_size,
#     dtype=str
# )

# main = []
# cols = ['tconst', 'nconst', 'category', 'ordering','titleId','language']

# # Only one loop, keeping all columns together
# for chunk in chunks:
#     movie = chunk[cols]
#     main.append(movie)

# # Concatenate all parts
# m4 = pd.concat(main, ignore_index=True)
# print(m4.head())
input_file = 'title.principals.tsv'
output_file = 'principal_tiny.tsv'

wanted_cols = ['tconst','nconst','ordering','category']  # Adjust for other files as needed

# Step 1: Find column index mapping
with open(input_file, 'r', encoding='utf-8') as f:
    header = f.readline().strip().split('\t')
col_idx = [header.index(col) for col in wanted_cols]

# Step 2: Extract just those columns line by line (super low RAM!)
with open(input_file, 'r', encoding='utf-8') as f_in, open(output_file, 'w', encoding='utf-8') as f_out:
    for line in f_in:
        fields = line.strip().split('\t')
        try:
            selected = [fields[i] for i in col_idx]
        except IndexError:
            continue  # skip broken/malformed lines
        f_out.write('\t'.join(selected) + '\n')

print("Columns written to", output_file)
```

```python
df = pd.read_csv(
    'principal_tiny.tsv',
    sep='\t',
    names=wanted_cols,
    dtype=str,            # Makes all columns string type, safe for mixed data
    low_memory=False      # Speeds up reading and prevents DtypeWarning
)
```

```python
df.to_csv('updated_principal.csv')
```

```python
import wikipedia
import webbrowser

# --- Configuration ---
PAGE_TITLE = "Kal Ho Naa Ho"
# ---------------------

try:
    # 1. Find the Wikipedia page
    print(f"Searching for Wikipedia page: '{PAGE_TITLE}'...")
    page = wikipedia.page(PAGE_TITLE, auto_suggest=False)
    print(f"Found page: {page.title}")

    # 2. Find the poster image URL
    poster_url = None
    for image_url in page.images:
        if "poster" in image_url.lower():
            poster_url = image_url
            break

    # Fallback: first available image
    if poster_url is None:
        if page.images:
            print("Could not find an image with 'poster' in the name.")
            print("Using the first available image as fallback...")
            poster_url = page.images[0]
        else:
            print("Error: No images found on this page.")
            exit()

    print(f"Found image URL: {poster_url}")

    # 3. Open image URL in default web browser
    print("Opening image URL in your default web browser...")
    webbrowser.open(poster_url)

except wikipedia.exceptions.PageError:
    print(f"Error: Page '{PAGE_TITLE}' not found.")
except wikipedia.exceptions.DisambiguationError as e:
    print(f"Error: '{PAGE_TITLE}' is a disambiguation page. Please be more specific.")
    print(f"Options: {e.options}")
except Exception as e:
    print(f"An error occurred: {e}")
```
